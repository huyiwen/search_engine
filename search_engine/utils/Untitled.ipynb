```python
from bs4 import BeautifulSoup, Tag
import lxml
```

```python
with open('saved_pages_bak/77.htm', 'r', encoding='utf-8') as f:
    html = f.read()
soup = BeautifulSoup(html, 'lxml')
```

```python
from Day1 import get_all_links
from pipe import select, where, chain
from url_normalize import url_normalize
from urllib.parse import urljoin
#list(get_all_links(html, link))

link = 'http://hqjt.ruc.edu.cn/jtgk/jgsz/index.htm'
raw_links = soup.find_all('a', href=True) | select(lambda x: x.attrs.get('href'))
```

```python
abs_links = raw_links | where(lambda x: not x.startswith('http')) | select(lambda x: url_normalize(urljoin(link, x)))
```
